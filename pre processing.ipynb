# PRE-PROCESSING

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/MachineLearnia/Python-Machine-Learning/master/Dataset/dataset.csv'
data = pd.read_csv(url, index_col=0, encoding = "ISO-8859-1")
data.head()
```

```python
df = data.copy()
df.head()
```

## from the EDA (exploration data analysis) : 
### keep the variables, selected before, identified as usefull :

```python
# recreate the 3 subsets we created before:

# rates of missing values by col
missing_rate = df.isna().sum()/df.shape[0]
# filter the blood's types and viral'stypes col
blood_columns = list(df.columns[(missing_rate < 0.9) & (missing_rate >0.88)])
viral_columns = list(df.columns[(missing_rate < 0.80) & (missing_rate > 0.75)])
# important cols :
key_columns = ['Patient age quantile', 'SARS-Cov-2 exam result']
```

```python
# create a filter with those 3 subsets for our df
df = df[key_columns + blood_columns + viral_columns]
df.head()
```

## TrainTest - Cleaning - Encoding

```python
from sklearn.model_selection import train_test_split
```

```python
# create a Train and Test set using scikit learn
# test size of 20% of our data set'size ; 
trainset, testset = train_test_split(df, test_size=0.2, random_state=0) # this funct allow us to split a df into a trainset and a testset randomly 
```

```python
# check the num of variables we got randomly in the two parts on our target col
trainset['SARS-Cov-2 exam result'].value_counts()
```

```python
testset['SARS-Cov-2 exam result'].value_counts()
```

## Enconding

```python
# create a dic that relies categories pos and neg to numbers
code = {'negative':0,
        'positive':1,
        'not_detected':0,
        'detected':1}
```

```python
# using pandas's map funct to apply this fun to all object's type col
def encoding(df):
    for col in df.select_dtypes('object').columns:
        df.loc[:,col] = df[col].map(code)
    return df
```

```python
df # we can see that all the qualitatives variables are replaced
```

```python
df.dtypes.value_counts() # to check that there are not object variables anymore in our df
```

```python
def imputation(df): 
    df = df.dropna(axis=0)
    return  df
```

```python
def feature_engineering(df):
    df['is sick'] = df[viral_columns].sum(axis=1) >= 1 # considered is sick not only for SARS-Cov-2 but for other diseases to
    df = df.drop(viral_columns, axis=1) # not importance variables
    return df
```

```python
def preprocessing(df):
    
    df = encoding(df)
    df = feature_engineering(df)
    df = imputation(df)
    # create our x and y to apply Machine Learning processes
    X = df.drop('SARS-Cov-2 exam result', axis=1) # we want to drop all the col
    y = df['SARS-Cov-2 exam result'] 
    
    print(y.value_counts()) # return the num of pos and neg cases we get after we applied the previous functions wrote before
    
    return X, y # we completed our pre processing function
```

```python
X_train, y_train = preprocessing(trainset) # on the train set
```

```python
X_test, y_test = preprocessing(testset) # on the test set
```

## Modelization

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline # for select K best technique
from sklearn.feature_selection import SelectKBest, f_classif # f_classif refeers to anova test
from sklearn.preprocessing import PolynomialFeatures
from sklearn.decomposition import PCA
```

```python
model = DecisionTreeClassifier(random_state=0) # random_state = 0 very important to keep a repetition in our results to evaluate the ideas with by the same way
model_1 = RandomForestClassifier(random_state=0)
model_2 = make_pipeline(PolynomialFeatures(2), SelectKBest(f_classif, k=10),RandomForestClassifier(random_state=0)) 
```

## Evaluation procedure

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report # confusion_matrix to check the relations between "false true" and "false false"; classification_report to make a conclusion of the precision and accuraty
from sklearn.model_selection import learning_curve # to understand if our model is on overfitting or under fitting
```

```python
def evaluation(model):
    
    model.fit(X_train, y_train) # start by train the model on the model X_train and y_train
    ypred = model.predict(X_test) # calculate the predictions of our model on the test set
    
    print(confusion_matrix(y_test, ypred)) # display the confusion matrix between y_test and our predictions ypred
    print(classification_report(y_test, ypred)) # same for clssification report
    
    # to use learning curve that use cross validation
    N, train_score, val_score = learning_curve(model, X_train, y_train, cv=4, scoring='f1', train_sizes=np.linspace(0.1, 1, 10))
    
    #to display it in on a matplotlib graph
    plt.figure(figsize=(12, 8))
    plt.plot(N, train_score.mean(axis=1), label='train score') # N = num of samples that used to train the model
    plt.plot(N, val_score.mean(axis=1), label='validation score') # same on validation score
    plt.legend()
```

```python
evaluation(model) # where 0 is neg and 1 is pos 
# we can see that we are in an overfitting situation where the train set is getting 100% (working well) and the validation score is very low
```
